```python
import imageio as img
import numpy as np
import matplotlib.pyplot as plt

# 1. Mengambil gambar berwarna dan mengubah menjadi grayscale
path = "/content/foto aku.jpeg"  # Ganti dengan path gambar Anda
img_color = img.imread(path)

# Mengubah gambar berwarna (RGB) menjadi grayscale
img_gray = np.dot(img_color[..., :3], [0.2989, 0.5870, 0.1140])  # Menggunakan rumus luminance

# 2. Menghitung histogram dari gambar grayscale
hist_gray, bins = np.histogram(img_gray.flatten(), bins=256, range=[0, 256])

# 3. Plot histogram menggunakan Matplotlib
plt.figure(figsize=(10, 6))

# Menampilkan gambar grayscale
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Menyembunyikan sumbu

# Menampilkan histogram
plt.subplot(1, 2, 2)
plt.plot(hist_gray, color='black')
plt.xlim([0, 255])
plt.title('Histogram Gambar Grayscale')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Jumlah Piksel')

plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah total piksel
total_pixels = np.sum(hist_gray)

print(f'Jumlah total piksel pada gambar grayscale: {total_pixels}')
```

```python
# Mencari intensitas dengan jumlah piksel tertinggi
dominant_intensity = np.argmax(hist_gray)
dominant_count = hist_gray[dominant_intensity]

print(f'Intensitas yang dominan: {dominant_intensity} dengan jumlah piksel: {dominant_count}')
```
